# Contexte de l'étude

Le livrable est un prototype d’un produit IA permettant de prédire le sentiment associé à un tweet.

Pour cela, nous allons recourir à des techniques d'analyse des sentiments des tweets. Cette méthode d'analyse repose sur des processus d’interprétation et de classification des émotions, des opinions et des jugements dans les données textuelles. Grâce à l'explosion des médias sociaux, les entreprises disposent de très nombreuses données pour identifier les sentiments des clients vis-à-vis de leurs prestations. 

Twitter compte en France près de 5 millions d'utilisateurs uniques par jour, près de 200 millions dans le monde. C'est une source cruciale d'informations pour la détection des clients insatisfaits. Il est devenu capital de pouvoir classifier un tweet comme positif, neutre ou négatif.

Afin de pouvoir optimiser et présenter le prototype, nous nous serviront d'une base de données Sentiment140 qui compte 1.6 millions de tweets. Les tweets ont soit une étiquette positive, soit une étiquette négative.
L'idée générale du projet est de créer un modèle qui envoie un tweet et récupère la prédiction de sentiment.

## Transformation du corpus

### Prétraitement avec la technique de la lemmatization

Maintenant que nous avons fait un premier nettoyage des tweets, nous allons faire une transformation plus approfondie de notre corpus.

Nous allons d'abord découper chaque phrase, pour qu'un avis soit une liste de phrases. C'est une méthode de segmentation pour laquelle nous allons utiliser le toolkit nltk.

Une fois les phrases séparées les unes des autres, nous effectuons la même opération avec les mots. La fonction nltk.RegexpTokenizer nous donne l'opportunité de ne garder que les caractères alphanumériques. Par ce biais, nous nous débarrassons de la ponctuation.

Il est important de faire l'étape de POS_tag AVANT la lemmatisation pour ne pas avoir de problème de surgénéralisation du sens des mots. L'étape d'étiquetage morpho-syntaxique permet d'associer un mot à ses informations grammaticales. Nous nous en servirons pour faire un tri des informations grammaticales pertinentes pour la recherche de sujets.

La lemmatisation consiste à ne garder uniquement que les formes canoniques d'un mot (grandes devient grand, détesté devient déteste) alors que la racinisation ne conserve que la racine du mot (optimiser devient optim). Dans cette première partie des transformations, nous faisons le choix de la lemmatisation. Aussi, l'algorithme fait attention à prendre en compte le type grammatical de certains mots (les noms, les verbes, les adjectifs et les adverbes) pour ne pas confondre un mot avec un autre.

La librairie nltk nous permet de trouver les stopwords d'une langue. Les stopwords sont des mots qui n'ont pas de valeur informative. Nous allons donc les déterminer et les éliminer de notre corpus de texte.

Attention à ne pas supprimer les stopwords qui auraient du sens dans notre étude. En effet, la phrase "I don't like this company" deviendrait "Do like company" et serait indiscernable d'un tweet de départ "I like this company". Nous allons donc choisir nous-mêmes la liste des stopwords.

### Prétraitement avec la technique de la racinisation (stemming)

Dans un deuxième temps, nous allons effectuer les mêmes opérations de transformation du corpus, mais en échangeant la lemmatisation avec une racinisation (stemming).

```python
ps = PorterStemmer()
tokenizer = nltk.RegexpTokenizer(r'\w+')
```

```python
def var_stemming(df, variable, stop_words, ps, tokenizer):
    
    
    
    resultat = [
        " ".join([
            ps.stem(word) for word in tokenizer.tokenize(texte)
            if not word in stop_words
        ]) for texte in df['tweet_nett']
    ]
    
    return resultat
```

```python
%%time
X_1['tweet_stem'] = var_stemming(X_1, 'tweet_nett', stop_words, ps, tokenizer)
```

```python
%%time
X_2['tweet_stem'] = var_stemming(X_2, 'tweet_nett', stop_words, ps, tokenizer)
```

Nous remarquons que l'opération est beaucoup plus rapide. En effet, l'étape de lemmatisation est plus lente, car elle demande à l'algorithme à faire attention aux POS_TAG des mots. Elle sera plus précise mais plus chronophage. Il faudra prendre ceci en compte quand nous ferons le bilan de l'impact de chaque méthode sur les prédictions grâce aux modèles.
